# Rolling a six-sided die where Random variable is sqrt(number on die)
 The sample space is {1, 2, 3, 4, 5, 6} and the probability of each independent outcome is 1/6, all equally likely. Since there are 6 possible outcomes, the total number of possible events (subsets) is: $2^{6} = 64$. 

```python
n = 5000
rolls = np.random.randint(1, 7, n) # 1-6
X = np.sqrt(rolls) # basically saying x = sqrt(number on die)

values = np.unique(X)
counts = np.unique(X, return_counts=True)[1] # returns return count which is second element of tuple
pmf = counts / n

```

```python
plt.figure(figsize=(7, 4))
plt.bar(values, pmf, width=0.15, edgecolor='black')
plt.xlabel("X = sqrt(pips)")
plt.ylabel("Empirical Probability")
plt.title("Empirical Mass Function of X (5000 rolls)")
plt.show() # all columns the same height which makes sense
```

```python
sorted_X = np.sort(X)
ECDF = np.arange(1, len(X)+1) / len(X) # in this case len(X) is n = 5000
plt.step(sorted_X, ECDF)
plt.title("ECDF of √(Die Roll)")
plt.xlabel("X = sqrt(pips)")
plt.ylabel("ECDF - Cumulative Probability")

plt.tight_layout()
plt.show()
```

# Rolling two six-sided die where Random variable is sum(number on die)
The sample space is The sample space is {1, 2, 3, 4, 5, 6}. We can see that because we are using two dice, it is a join sample space. Since each die has 6 faces, there are 36 total possible outcomes in rolling those two dice. We can see that there are 11 possible sum combinations, where each is not equally likely. This is because when we roll "7", we can get a 2+5, or 4+3 versus a "2", we can only get a 1+1. Since there are 11 possible sums, the number of possible events is $2^{11} = 2048$.

```python
n = 10000
die1 = np.random.randint(1, 7, n) 
die2 = np.random.randint(1, 7, n)# 1-6
X = die1 + die2 

values = np.unique(X)
counts = np.unique(X, return_counts=True)[1] 
pmf = counts / n
```

```python
plt.figure(figsize=(7, 4))
plt.bar(values, pmf, width=0.15, edgecolor='black')
plt.xlabel("Sum of Two Die")
plt.ylabel("Empirical Probability")
plt.title("Empirical Mass Function of X (10000 rolls)")
plt.show() 
```

```python
sorted_X = np.sort(X)
ECDF = np.arange(1, len(X)+1) / len(X) # in this case len(X) is n = 10000
plt.step(sorted_X, ECDF)
plt.title("Empirical Cumulative Distribution Function (ECDF)")
plt.xlabel("Sum of Two Die")
plt.ylabel("ECDF - Cumulative Probability")

plt.tight_layout()
plt.show()
```

You can use https://docs.scipy.org/doc/scipy/reference/stats.html to generate values for the pdf/cdf and generate samples of random variates.


- Logistic distribution (similar to normal): 
$$
F(x; \sigma) = \dfrac{1}{1+e^{-x/\sigma}} 
$$
with $\sigma >0$.
- Exponential distribution (similar to log-normal):
$$
F_X(x) = \begin{cases}
0, & x<0 \\
1 - e^{-\lambda x}, & x \ge 0,
\end{cases}
$$
with $\lambda > 0$.
- Negative Binomial (similar to Poisson): The probability mass function for positive integers is:
$$
f(k;r,p) = \dfrac{(k+r-1)!}{k!(r-1)!}(1-p)^k p^r, \quad \text{ for $k=0,...n$}
$$
You can interpret this as follows: Flip a coin that comes up heads with probability $p$ until you get $r$ heads, and then stop. What is the probability of stopping at each $k=0,1,2,...$?
- Categorical (similar to Bernoulli): The probability mass function over $k = 1, 2, ..., K$ categories is
$$
f(k;p_1,...,p_K) = p_1^{k=1}p_2^{k=2}...p_K^{k=K}
$$
where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.

# Logistic Distribution

The support of the logistic distribution is from -∞ to ∞ as there is no value where 
e^(-x/0) is less than 0. This means support is all real numbers in this case


```python
sigmas = [0.5, 1, 2]
support = np.linspace(-10, 10, 1000)
for s in sigmas:
    pdf = logistic.pdf(support, loc=0, scale=s)
    plt.plot(support, pdf, label=f'σ={s}')
plt.title('Logistic Distribution PDF for different σ')
plt.ylabel('Density')
plt.legend()
plt.show()

for s in sigmas:
    cdf = logistic.cdf(support, loc=0, scale=s)
    plt.plot(support, cdf, label=f'σ={s}')
plt.title('Logistic Distribution CDF for different σ')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()
```

```python
sigma = 1
n = 1000
rng = np.random.default_rng(0)
sample = logistic.rvs(loc=0, scale=sigma, size=n, random_state=rng)

sns.kdeplot(sample, label='Empirical KDE', fill=True)
plt.plot(support, logistic.pdf(support, loc=0, scale=sigma), color='C1', lw=2, label='Theoretical PDF')
plt.title(f'Logistic PDF (σ={sigma})')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.show()


x_sorted = np.sort(sample)
y_ecdf = np.arange(1, n+1) / n
plt.step(x_sorted, y_ecdf, where='post', label='Empirical ECDF')
plt.plot(support, logistic.cdf(support, loc=0, scale=sigma), color='C1', lw=2, label='Theoretical CDF')
plt.title(f'Logistic CDF (σ={sigma})')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()
```

# Negative Binomial Distribution
the support for the negative binomial distribution is all integers greater than or equal to 0.

```python
success = 10
p_values = [0.1, 0.2, 0.5, 0.9]  
x = np.arange(0, 50)  

plt.figure(figsize=(10, 6))
for p in p_values:
    pmf = nbinom.pmf(x, success, p)
    plt.scatter(x, pmf, marker='o', label=f'p={p}')

plt.xlabel('Number of failures (k)')
plt.ylabel('Probability')
plt.title(f'Negative Binomial PMF for various p values')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Here, I Plot CDF for different p values
plt.figure(figsize=(10, 6))
for p in p_values:
    cdf = nbinom.cdf(x, success, p)
    plt.step(x, cdf, where='post', label=f'p={p}')

plt.xlabel('Number of failures (k)')
plt.ylabel('Cumulative Probability')
plt.title(f'Negative Binomial CDF for various p value')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
n = 10
p = 0.5
size = 1000
rng = np.random.default_rng(0)
sample = nbinom.rvs(n, p, size=size, random_state=rng)

x = np.arange(0, 30)
theoretical_pmf = nbinom.pmf(x, n, p)
sns.kdeplot(sample, label='Empirical')
plt.plot(x, theoretical_pmf, 'r-', label='Theoretical PMF')
plt.title('Negative Binomial PMF Comparison')
plt.xlabel('Number of failures')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
x_sorted = np.sort(sample)
y_ecdf = np.arange(1, len(sample) + 1) / len(sample)
theoretical_cdf = nbinom.cdf(x, n, p)
plt.step(x_sorted, y_ecdf, label='Empirical CDF')
plt.plot(x, theoretical_cdf, 'r-', label='Theoretical CDF')
plt.title('Negative Binomial CDF Comparison')
plt.xlabel('Number of failures')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()
```
